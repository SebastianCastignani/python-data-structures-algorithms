### **Ejercicio 3**

Escribir una función que calcule y retorne el factorial de un número natural. La operación factorial (!) se define de la siguiente manera:

image.png

```
def factorial(n):  # <-- acá vos le pusiste ese nombre
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)  # <-- acá la función se llama a sí misma
print(factorial(5))
```

### **Ejercicio 11**

Escribir un programa que pida dos años y escriba cuántos años bisiestos hay entre esas dos fechas (incluidos los dos años).

```
def es_bisiesto(ano):
    return (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0)

# Pedimos los años al usuario
inicio = int(input("Ingresa el primer año: "))
fin = int(input("Ingresa el segundo año: "))

# Nos aseguramos de que inicio sea menor que fin
if inicio > fin:
    inicio, fin = fin, inicio

# Contamos los años bisiestos
contador = 0
for ano in range(inicio, fin + 1):
    if es_bisiesto(ano):
        contador += 1

print(f"Hay {contador} años bisiestos entre {inicio} y {fin} (incluidos).")
```

### **Ejercicio 12**

Escribir una función que tome por parámetro un numero entero y retorne la suma de sus dígitos. Para obtener los digitos deben ir dividiendo (division entera) por 10 sucesivamente y tomando el resto (digito que esta mas a la derecha).

```
def digitos(numero):
    suma = 0
    while numero != 0:
        suma += numero % 10  # agarro el último dígito
        numero = numero // 10  # Elimino el último dígito
    return suma


digitos(15)
print(1 // 10 )
```


### **Ejercicio 13**

Escribir un programa que pida números positivos (Debe dejar de pedir números cuando se ingrese uno negativo). Al finalizar debe mostrar la cantidad de números cuya sumatoria de dígitos fue menor que 10. Utilizar una o más funciones, según sea necesario.

```
def procesar_numeros():
    numero = 0
    contador = 0

    while numero >= 0:
        numero = int(input("Ingrese un número positivo (negativo para terminar): "))
        if numero >= 0 and digitos(numero) < 10:
            contador += 1

    print("Cantidad de números cuya suma de dígitos fue menor que 10:", contador)

procesar_numeros()
```

### **Ejercicio 14**

Escribir un programa que solicita el ingreso de números primos. La lectura finalizará cuando ingrese un número que no sea primo. Al finalizar el programa, mostrar el factorial del número que tenga la mayor suma de dígitos entre todos los ingresados. Utilizar una o más funciones, según sea necesario.

```
#esPrimo:int -> bool
#sumaDigitos:int -> int
#factorial:int -> int
#mayor:int:int -> int
def esPrimo(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True


def procesar_primos():
    num = int(input("Ingrese un número primo: "))

    max_suma = 0
    num_con_max_suma = 0

    while esPrimo(num):
        suma = digitos(num)
        if suma > max_suma:
            max_suma = suma
            num_con_max_suma = num
        num = int(input("Ingrese un número primo: "))

    print("El número con mayor suma de dígitos fue:", num_con_max_suma)
    print("Su factorial es:", factorial(num_con_max_suma))

procesar_primos()
```
