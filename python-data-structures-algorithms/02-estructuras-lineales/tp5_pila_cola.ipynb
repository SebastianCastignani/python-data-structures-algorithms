## **Parte 1: Pilas dinámicas**

### **Ejercicio 1**

Implementar el TDA Pila (Stack), con las siguientes operaciones:
- Crear (__init__())
- Vaciar
- Apilar elemento (push)
- Desapilar elemento (pop)
- Obtener primer elemento (top)
- Obtener tamaño de pila
- Clonar
- Esta vacía.
- __repr__(). Para poder imprimir una Pila por consola

```
import copy as cp # paquete para no tener problemas con las estructuras que tienen multiples niveles

class Pila:
  def __init__(self, listaInicial=None):
    self.pila = []
    if listaInicial:
      self.pila = cp.deepcopy(listaInicial)

  def tamanio(self)->int:
    return len(self.pila)

  def estaVacia(self)->bool:
    return self.tamanio() == 0

  def apilar(self, dato:int)->None:
    self.pila.append(dato)

  def tope(self)->int:
    datoCima = None
    if not self.estaVacia():
      datoCima = self.pila[len(self.pila)-1] # leo ultimo elemento
    return datoCima

  def desapilar(self)->int:
    dato = None
    if not self.estaVacia():
      dato = self.pila.pop()
    return dato

  def vaciar(self)->None:
    self.pila = []

  def clonar(self): # retorna una pila que contengan los mismos elementos pero en otra posicion de memoria
    return cp.deepcopy(self)

  def __repr__(self)->str:
    return str(self.pila)+"<->"
```

```
pila1 = Pila()
print(pila1)
pila1.apilar(3);pila1.apilar(2);pila1.apilar(7);pila1.apilar(4)
print(pila1)
print(pila1.tope())
print(pila1.desapilar())
print(pila1)
print(pila1.tope())
pila2 = pila1.clonar()
print("pila2",pila2)
```

Escribir una función que invierta el orden de una pila. No debe devolver una nueva pila invertida, sino invertir la pila que ingresa por parámetro.

```
def invertirPila(pila):
  pilaAux = pila.clonar()
  pila.vaciar()
  while not pilaAux.estaVacia():
        pila.apilar(pilaAux.desapilar())
  return pila


pilaEntero = Pila()
pilaEntero.apilar(1)
pilaEntero.apilar(2)
pilaEntero.apilar(3)
pilaEntero.apilar(4)

print(invertirPila(pilaEntero))
```

### **Ejercicio 6**

Escribir una función que elimine de una pila todas las ocurrencias de un elemento dado. Usar una pila auxiliar.

```
def eliminarOcurrencias(unaPila:Pila , unElementoAEliminar:int)->None: #Declaro mi función que recibe una pila y un elemento a eliminar
  pilaAux = Pila() #Declaro mi pila auxiliar
  while not unaPila.estaVacia(): #Mientras mi pila NO esté vacía
    esteElemento = unaPila.desapilar() #Desapilo. Pero Ojo! Me guardo el dato, porque lo voy a querer procesar. Quiro tomar decisiones mientras recorro
    if esteElemento != unElementoAEliminar: #Si el elemento que desapilé es DISTINTO del elemento a eliminar
      pilaAux.apilar(esteElemento) # apilo el elemento en mi pila auxiliar. Sino, no hago nada
  while not pilaAux.estaVacia(): #Ahora en pila auxiliar tengo solo los elementos que me interesan, así que desapilo y lo guardo como siempre. Mientras pila auxiliar NO esté vacía
    unaPila.apilar(pilaAux.desapilar()) #Apilo en mi pila los elementos que voy desapilando de la pila auxiliar, que solo contiene los datos que me interesan

miPila = Pila([1 , 2 , 3 , 4 , 5 , 5 , 4 , 3 , 2 , 1])
print(miPila)
eliminarOcurrencias(miPila , 3)
print(miPila)
```

### **Ejercicio 11**

Implementar el TDA Cola (Queue), con las siguientes operaciones:
- Crear (__init__())
- Vaciar
- Encolar elemento (enqueue)
- Desancolar elemento (dequeue)
- Obtener primer elemento (top)
- Obtener tamaño de cola
- Clonar
- Esta vacía.
- __repr__(). Para poder imprimir una Cola por consola

```
class Cola:
  def __init__(self):
    self.cola = []

  def tamaño(self)->int:
    return len(self.cola)

  def estaVacia(self)->bool:
    return self.tamaño() == 0

  def encolar(self, dato:int)->None:
    self.cola.append(dato) #Encolando por el final
    #self.cola.insert(0,dato) #Encolando por el inicio

  def tope(self)->int:
    datoCima = None
    if not self.estaVacia():
      datoCima = self.cola[0]  #Desencolo por el inicio
      #datoCima = self.cola[len(self.cola)-1] #Desencolo por el final
    return datoCima

  def desencolar(self)->int:
    dato = None
    if not self.estaVacia():
      dato = self.cola.pop(0)  #Desencolando por el inicio
      #dato = self.cola.pop() #Desencolando por el final
    return dato

  def vaciar(self)->None:
    self.cola = []

  def clonar(self):
    return cp.deepcopy(self)

  def __repr__(self)->str:
    return "<-"+str(self.cola)+"<-"
```

```
#colaOriginal = <-[2,3,1,5]<-
#cola = <-[5,1,3,2]<-
#pilaAux = []<->

def invertirCola(cola:Cola)->None: #Mi función invertir toma una cola y no devuelve nada, puese modifica la cola ingresada
  pilaAux = Pila() #Instancio una Pila como estructura de datos adiscional
  while not cola.estaVacia(): #Mientras aún haya elementos:
    pilaAux.apilar(cola.desencolar()) #Apilo los elementos que voy sacando de la fila
  while not pilaAux.estaVacia(): #Luego, mientras la pila no esté vacía
    cola.encolar(pilaAux.desapilar()) #Voy desapilando y encolando los elementos. Acá aprovechamos que desapilar ya nos invierte el orden de los elementos.

cola1 = Cola()
cola1.encolar(2);cola1.encolar(3);cola1.encolar(1);cola1.encolar(5);
print(cola1)
invertirCola(cola1)
print(cola1)
```

### **Ejercicio 16**

Escribir una función que recibe una cola de enteros y genera dos colas: una con los números primos y otra con los números no primos, de la cola de entrada. Al terminar la función, la cola de entrada no debe estar modificada.

```
def esPrimo(num): #Primero defino mi función "esPrimo" que toma un número
  if num == 2:
    return True
  for n in range(2,num):
    if num % n == 0:
      return False
    return True


def separaPrimos(cola:Cola)->tuple[Cola,Cola]:
  colaAux = cola.clonar()
  colaPrimos = Cola()
  colaCompuestos = Cola()
  while not colaAux.estaVacia():
    dato = colaAux.desencolar()
    if esPrimo(dato):
      colaPrimos.encolar(dato)
    else:
      colaCompuestos.encolar(dato)
  return colaPrimos,colaCompuestos

c1 = Cola()
c1.encolar(1);c1.encolar(3);c1.encolar(2);c1.encolar(5);c1.encolar(4);c1.encolar(2);c1.encolar(3);c1.encolar(4);
print(c1)
print(separaPrimos(c1))
```
